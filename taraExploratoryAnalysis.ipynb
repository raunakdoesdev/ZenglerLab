# Loading Dataset
Now that I've downloaded and extracted it. Let's load it into the pandas library and see what kind of data is contained.

```
dataset = pd.DataFrame.from_csv('TARA243.OG.profile.release', sep=r"\s*");
dataset.head(5)  # visualize first five rows
```

# Analysis of a Single COG

```
cog1 = dataset.iloc[1]
cog1.head(5) # print first five valeus in COG
```

```
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [75, 15]

import matplotlib.pyplot as plt
plt.xticks(rotation=90)  # vertical axis labels
plt.bar(cog1.axes[0], cog1)  # visaulize cog1 data in barchart
```
